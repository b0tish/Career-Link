```python
from bs4 import BeautifulSoup


def strip_html(text):
    return BeautifulSoup(text, "html.parser").get_text()

df["Experience"] = df["Resume_html"].apply(strip_html)
```

```python
import re

def normalize(text):
   
    text = text.lower()
   
    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)
   
    text = re.sub(r'\s+', ' ', text)
    return text.strip()



print("Columns in df:")
print(df.columns)


df['Resume_normalized'] = df['Experience'].apply(normalize)


df
```

```python
import re

def extract_experience_sentences(text):
    """Split text into sentences and filter those related to experience."""
    # Split by sentence endings
    sentences = re.split(r'[.!?]', text)
    # Filter for those that mention experience
    return [s.strip() for s in sentences if 'experience' in s.lower()]
    

# Apply this to create a new column with experience-related sentences
df['Experience_sentences'] = df['Experience'].apply(extract_experience_sentences)
```

```python
import re

# Strip out newlines and excess spaces first
def clean_newlines(text):
    return " ".join(text.split())  # This converts all whitespace (newlines, tabs, multiple spaces) into single spaces


# Split cleaned text into sentences
def extract_sentences(text):
    cleaned = clean_newlines(text)
    return re.split(r'(?<=[.!?]) +', cleaned)


# Apply this to your parsed Experience column
df['Experience_sentences'] = df['Experience'].apply(extract_sentences)


# Now view the first few to confirm
df.head(5)
```
